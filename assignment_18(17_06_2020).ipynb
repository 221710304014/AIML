Apply onehotencoder to a categorical column

```
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[('Encoder',OneHotEncoder(),['department'])],remainder="passthrough")

new_d=ct.fit_transform(data)
new_d=pd.DataFrame(new_d)
new_d=new_d.iloc[:,0:9]
new_d.columns=['dep_sales','dep_technical','dep_support','dep_product_mng','dep_marketing','dep_IT','dep_hr',
              'dep_accounting','dep_management']
new_d=pd.concat([data,new_d],axis=1)
new_d=new_d.drop('department',axis=1)
new_d
```

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
new_d.salary=le.fit_transform(new_d.salary)
new_d
```
